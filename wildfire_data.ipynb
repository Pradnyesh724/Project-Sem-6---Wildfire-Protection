```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
n_samples = 1000
```

```python
data = pd.DataFrame({
    "temperature": np.random.uniform(20, 50, n_samples),
    "NDVI": np.random.uniform(0.2, 0.8, n_samples),
    "humidity": np.random.uniform(20, 80, n_samples),
    "wind_speed": np.random.uniform(2, 25, n_samples),
    "slope": np.random.uniform(0, 10, n_samples),
})
```

```python
data["fire_risk"] = (
    (data["temperature"] > 35) &
    (data["NDVI"] < 0.4) &
    (data["humidity"] < 40) &
    (data["wind_speed"] > 10)
).astype(int)
```

```python
data.to_csv("wildfire_data.csv", index=False)
```
